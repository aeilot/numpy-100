#### 35. How to compute ((A+B)*(-A/2)) in place (without copy)? (★★☆)

````python
# To calculate in-place, ufuncs must be used with the out parameter.
# Otherwise, numpy creates temporary arrays for the calculation.

A = np.ones((5,5))
B = np.ones((5,5))

np.add(A,B,out=B)
np.divide(A,2,out=A)
np.multiply(A,-1,out=A)
np.divide(B,A,out=A)

print(A)
````

#### 52. Consider a random vector with shape (100,2) representing coordinates, find point by point distances (★★☆)

#### 53. How to convert a float (32 bits) array into an integer (32 bits) array in place?

````python
Z = (np.random.rand(10)*100).astype(np.float32)
# Create a new array 'Y' that shares the exact same memory buffer as 'Z'.
# However, 'Y' is instructed to interpret that binary data as 32-bit integers.
# (This is only possible because float32 and int32 both occupy 4 bytes).
# Binary Reinterpretation
Y = Z.view(np.int32)
# Copy the values from 'Z' into 'Y'.
# Y = Z but different interpretation (dtype)
# Converting in place
Y[:] = Z
print(Y)
````
